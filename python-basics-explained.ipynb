## A tuple is a collection of objects which ordered and immutable. Tuples are sequences, just like lists. The differences between tuples and lists are, the tuples cannot be changed unlike lists and tuples use parentheses, whereas lists use square brackets
1. any()	Returns True if any element present in a tuple and returns False if the tuple is empty
1. min()	Returns smallest element (Integer) of the Tuple
1. max()	Returns largest element (Integer) of the Tuple
1. len()	Returns the length of the Tuple
1. sorted()	Used to sort all the elements of the Tuple
1. sum()	Returns sum of all elements (Integers) of the Tuples

```python
# Creating a Tuple
tup1 = ('physics', 'chemistry', 1997, 2000);
tup2 = (1, 2, 3, 4, 5 );
print(tup1)
print(tup2)
###Combining 2 Tupples.
tup3=tup1 + tup2
print(tup3)
### Accessing items in tupple
print(tup1[0])
print(tup1[0:3])
print(tup1[::-1])

#Tuple Functions 
print(any(tup1))  ## Any is just to see tupple is empty or has element

## Min max function
print(f"In Tup2 the min value={min(tup2)} and maximum value is={max(tup2)} also length is:{len(tup2)} also sum={sum(tup2)}")

## sorted will in ascending order
tup4=(-1,23,5,8,0,22)
tup5=("indian","america","japan")
print(sorted(tup4))
print(sorted(tup5))  # Sorted is alphabaticall order only. 


# Import the 'namedtuple' function from the 'collections' module
from collections import namedtuple
# Set the tuple
class1 = namedtuple("Class", "name maths eng hindi")
sub = class1(name="Virat", maths=90,eng=92,hindi=89)
# Print the tuple
print(sub)
# Print each item in the tuple
print(sub.name,sub.maths,sub.eng,sub.hindi)

```

## Set is a collection which is unordered and unindexed. In Python, sets are written with curly brackets.
1. add()	Adds an element to the set
1. clear()	Removes all the elements from the set
1. copy()	Returns a copy of the set
1. difference()	Returns a set containing the difference between two or more sets
1. difference_update()	Removes the items in this set that are also included in another, specified set
1. discard()	Remove the specified item
1. intersection()	Returns a set, that is the intersection of two other sets
1. intersection_update()	Removes the items in this set that are not present in other, specified set(s)
1. isdisjoint()	Returns whether two sets have a intersection or not
1. issubset()	Returns whether another set contains this set or not
1. issuperset()	Returns whether this set contains another set or not
1. pop()	Removes an element from the set
1. remove()	Removes the specified element
1. symmetric_difference()	Returns a set with the symmetric differences of two sets
1. symmetric_difference_update()	inserts the symmetric differences from this set and another
1. union()	Return a set containing the union of sets
1. update()	Update the set with the union of this set and others


```python
set1 = {1,2,3,4}
for x in set1:print(x)
#thisset[0] ## This raises an error and set can't accessed using index
set1.add("orange")
print(set1)
set1.update(["orange", "mango", "grapes"])
print(set1)
print(len(set1))
### Remove item from set
set1.remove("mango")
print(set1)

#Pop Method in sets remove the last itemsRemember that sets are unordered, 
#so you will not know what item that gets removed.
set1.pop()
print(set1)

########Updates in sets , please note set ignores duplicates
set2 = {"a", "b" , "c"}
set3 = {1, 2, 3,'a'}

set2.update(set3)
print(set2)

########clear set 3 
print(set3)
set3.clear()
print(set3)

## Copy Values 
set4=set2.copy()
print(set4)
set5={1,2,3,4,5}
set6={4,5,6,7,8}
print("Difference ",set5.difference(set6))

### Difference update
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
print("Difference 2",x.difference_update(y))

## Discard Method 
fruits = {"apple", "banana", "cherry"}
fruits.discard("jamatadi") ## This does not raise error if items does not exist 
print(fruits)

### Union of Set
a={1,2,3,4,5}
b={3,4,6,7}
print(a.union(b))
```

```python
 # To loop over two or more sequences at the same time, the entries can be paired with the zip() function.

num = [1, 2, 3]
colors= ['red', 'blue', 'green']
for i in zip(num, colors):
    print(i)

num = [1, 2, 3,4,5,6,7,8]  ## Output here would be the Same as Zip will combine only 3 items
colors= ['red', 'blue', 'green']
for i in zip(num, colors):
    print(i)

```

```python
#Make an iterator that aggregates elements from each of the iterables. 
#If the iterables are of uneven length, missing values are filled-in with fillvalue. 
#Iteration continues until the longest iterable is exhausted.
from itertools import zip_longest
colors=['red','apple','three']  
num=[1,2,3,4,5]
for i in zip_longest(colors,num):# Unlike Zip Zip longest will combine the non null values with None
    print (i)

from itertools import zip_longest
colors=['red','apple','three']
num=[1,2,3,4,5]
for i in zip_longest(colors,num,fillvalue='NoMatch'):# Here the Non existing is replaced with NoMatch
    print (i)
```

```python
# Looping through the sequence (list) in sorted order(ascending order) using sorted () function.
num=[1,10,3,-1,4,34]
for i in sorted(num):  # Print the list in ascending order
    print (i,end=" ")   # End is written to print the output in Ascending order.
print()
# Looping through the sequence (list) in sorted order(descending order) using sorted () function.
list1=[1,10,3,-1,4,34]
for i in sorted(list1,reverse=True):  # Print the list in descending order reverse=True is for that
    print (i,end=" ") # End is written to print the output in Ascending order.
```

```python
# Looping Through Dicitionary 
base=dict(a=1,b=2,c=3,d=4,e=5,f=6)  # Defining Dictionary dict word we dont need to put quote for similar types of values
for key,values in sorted(base.items()):  # the dictionary is printed in sorted order ascending 
    print(key,values,end=" ")

print()
  
base=dict(a=1,b=2,c=3,d=4,e=5,f=6)  # Defining Dictionary dict word we dont need to put quote for similar types of values
for key,values in sorted(base.items(),reverse=True):  # the dictionary is printed in sorted order reverse 
    print(key,values,end=" ")
```
